2. Ejercicios intervalo de confianza: Para cada uno de los ejercicios interpreta los resultados y piensa que ventaja a nivel de negocio podemos sacar de calcularlo.

```python
# Trabajas para una plataforma de comercio en línea y deseas determinar el intervalo de confianza al 95% para el precio promedio de productos con una cantidad de media de valoraciones al producto (mean_product_ratings_count) superior a 500.

    ## Creamos un dataframe para los productos que tienen más de 500 calificaciones:

calif_mas_500 = df[df["mean_product_ratings_count"] > 500]

calif_mas_500
```

```python
## Intervalo de confianza =  Estimación Puntual ± Valor Crítico * Error Estándar
# lo primero que tenemos que hacer es calcular la estimación puntual, en este caso calcularemos la media de la duración de las llamadas
media_precio_promedio = calif_mas_500["price"].mean()
# después vamos a calcular el error estándar utilizando el método "sem" de la librería stats
error_precio_promedio = stats.sem(calif_mas_500["price"])
# definimos el nivel de confianza (95% en este caso)
nivel_confianza = 0.95
# calculamos los grados de libertad de la muestra. Recordad que debemos restar el total de datos que tenemos -1. 
grados_libertad = len(calif_mas_500["price"]) - 1
# calculamos el valor crítico de la muestra
valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df=grados_libertad)
# calculamos el intervalo de confianza

limite_inferior_precio_promedio = media_precio_promedio - valor_critico * error_precio_promedio
limite_superior_precio_promedio = media_precio_promedio + valor_critico * error_precio_promedio

print(f"El intervalo de confianza del 95% para la variable 'price' es: {limite_inferior_precio_promedio}, {limite_superior_precio_promedio}")
```

```python
# Eres un analista de ventas en una plataforma de comercio electrónico y quieres calcular el intervalo de confianza al 90% para el número promedio de unidades vendidas(mean_units_sold_per_product) por producto.

## Intervalo de confianza =  Estimación Puntual ± Valor Crítico * Error Estándar
# lo primero que tenemos que hacer es calcular la estimación puntual, en este caso calcularemos la media de la duración de las llamadas
media_unidades_promedio = df["mean_units_sold_per_product"].mean()
# después vamos a calcular el error estándar utilizando el método "sem" de la librería stats
error_unidades_promedio = stats.sem(df["mean_units_sold_per_product"])
# definimos el nivel de confianza (90% en este caso)
nivel_confianza = 0.90
# calculamos los grados de libertad de la muestra. Recordad que debemos restar el total de datos que tenemos -1. 
grados_libertad = len(df["mean_units_sold_per_product"]) - 1
# calculamos el valor crítico de la muestra
valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df=grados_libertad)
# calculamos el intervalo de confianza

limite_inferior = media_unidades_promedio - valor_critico * error_unidades_promedio
limite_superior = media_unidades_promedio + valor_critico * error_unidades_promedio

print(f"El intervalo de confianza del 95% para la variable 'mean_units_sold_per_product' es: {limite_inferior}, {limite_superior}")
```
