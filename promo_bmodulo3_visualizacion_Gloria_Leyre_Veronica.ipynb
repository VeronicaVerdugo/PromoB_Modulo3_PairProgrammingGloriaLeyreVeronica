```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("files/M3_24.08_merchant_fill_nonulls.csv")
df.head()
```

```python
# Ejercicios
# Contesta a las siguientes preguntas utilizando la visualización más apropiada.
# 1. ¿Cómo se distribuyen los precios de los productos?
sns.histplot(data=df, x= "price", color = "blue", bins = 20, kde = True);

```

```python
df.info()
```

```python
# 2. ¿Cuál es la relación entre el precio y la calificación promedio de los productos?
# Hacemos REGRESIÓN que sirve para idenrtificar la relación entre dos variables
sns.regplot(data=df, 
            x="mean_product_ratings_count",
            y="price",
            marker = "x",
            line_kws= {"color": "violet", "linewidth":3},
            scatter_kws= {"color":"blue"});
```

```python
# 3. ¿Qué porcentaje de productos tiene el distintivo "Producto local"?
#  badge_local_product
# Creamos variable para sacar el total de cada valor dentro de la columna de productos locales.
# El reset index es para dejar la misma estructura del Data Frame

producto_local = df["badge_local_product"].value_counts().reset_index() 
print (producto_local)

# Los colores no nos gustan, los cambiamos
colores = ["grey","turquoise"]

plt.pie( "count",
        data = producto_local,
        labels= "badge_local_product",
        colors = colores,
        autopct = "%1.1f%%");

# Hacemos una visualización PIEPLOT desde la columna numérica COUNT etiquetada con los valores en STRING (Si/NO)
```

```python
# 4. ¿Qué porcentaje de productos tiene el distintivo "Buena calidad del producto":
# badge_product_quality
producto_buena_calidad = df["badge_product_quality"].value_counts().reset_index() 
print (producto_buena_calidad)
colores = [ "cadetblue", "teal"]

plt.pie( "count",
        data = producto_buena_calidad,
        labels= "badge_product_quality",
        colors = colores,
        autopct = "%1.1f%%");
```
