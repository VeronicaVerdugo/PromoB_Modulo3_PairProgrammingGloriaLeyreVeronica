```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_rating = pd.read_csv("files/rating-and-performance.csv",index_col= 0)
df_insight = pd.read_csv("files/insight.csv", index_col= 0)
```

```python
df_rating.head()
```

```python
df_insight.head()
```

```python
# EJERCICIOS: 
# 1. Unid los dos DataFrames en uno solo, usando el método que considereis más correcto. 
# Querremos quedarnos con todos los productos que estén en los dos conjuntos de datos.
df_unido = df_rating.merge(df_insight, left_on = "merchant_id", right_on = "merchant.id")
df_unido.head()
```

```python
# 2. Explorad el nombre de las columnas, ¿hay alguna que esté repetida o que no nos aporte información de valor? 
# ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. 
# NOTA Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo tiene un valor.

df_unido.drop(columns=['merchant.id'], inplace=True)
df_unido.head()
```

```python
df_unido.duplicated
```

```python
df_unido.drop(columns=['merchant.ratings.count'], inplace=True)
df_unido.head()
```

```python
df_unido.drop(columns=['title'], inplace=True)
df_unido.head()
```

```python
# 3. Algunas de las columnas tienen . en sus nombres y en otras tenemos _. 
# Unifica el nombre de las columnas para que todas tengan "_". Además, todas deben estar en minúsculas.

col_nuevas = {col : col.lower().replace(".", "_") for col in df_unido.columns}
col_nuevas
df_unido.rename(columns=col_nuevas, inplace=True)
df_unido.head()
```

```python
# En la columna de merchant_info_subtitle, si nos fijamos hay algunas celdas donde tenemos entre paréntesis el número total de calificaciones para el comerciante 
# (que es la misma información que la que tenemos en la columna "merchant_rating_count"). 
# Eliminad de la columna "merchant_info_subtitle" la parte del string de los paréntesis. 
# Guardar el resultado de vuestra limpieza en una columna que se llame igual, es decir, "merchant_info_subtitle".

df_unido[["merchant_info_subtitle"]] = df_unido["merchant_info_subtitle"].str.split("(", expand=True).get([0])
## el expand=True rompe la lista en columnas
## str sirve para hacerlo en las series
```

```python
# 5. Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, 
# y ahora tenemos un string vacío, lo cual no tiene mucho sentido. 
# En este ejercicio debereis reemplazar esos strings vacios por nulos de Pandas 
# (NOTA estos deben ser indicados de la siguiente forma np.nan, y deberéis importar numpy).
import numpy as np
```

```python
df_unido.replace('', np.nan, inplace=True)
```

```python
#6. Hay ciertas columnas que no vamos a necesitar para nuestro análisis. 
# En concreto, vamos a eliminar las columnas de badges_count, product_variation_inventory, urgency_text, merchant_title, merchant_has_profile_picture, merchant_profile_picture, 
# product_picture, total_urgency_count, has_urgency_banner y urgency_text_rate. Guardad los cambios realizados hasta este punto en un csv para usarlo en los ejercicios de la lección de mañana.

df_unido.drop(columns=["badges_count", "product_variation_inventory", "urgency_text", "merchant_title", "merchant_has_profile_picture", "merchant_profile_picture", "product_picture", "total_urgency_count", "has_urgency_banner", "urgency_text_rate"], inplace=True)
df_unido
```

```python
# 7. Pongamos en práctica el filtrado de datos en Pandas:

#7.1 Seleccionad aquellos productos cuyo color es verde. ¿Cuántos productos tenemos?
df_green = df_unido[df_unido["product_color"].str.contains("green", regex = True, na = False)]
df_green
# 7.2 Seleccionad aquellos productos cuyo color es verde y rojo. ¿Cuántos productos tenemos de cada tipo de color?
df_green_and_red = df_unido[df_unido["product_color"].str.contains("red|green" , regex = True, na = False)]
df_green_and_red
```

```python
# 7.3 Seleccionad aquellos productos de la talla M, que valgan más de 15. ¿Cuántos productos tenemos?
df_unido_M = df_unido.loc[(df_unido["product_variation_size_id"]=="M")& (df_unido["price"]>15.00) , :]
df_unido_M.shape
# 17 PRODUCTOS
```

```python
# 7.4 Seleccionad aquellos productos cuyo precio este entre 1 y 2 (ambos inclusive). ¿Cuántos productos tenemos?

df_unido_precio = df_unido[df_unido["price"].between(1, 2, inclusive="both")].head()
df_unido_precio.shape
# La respuesta es 5 productos.
```

```python
# 7.5 Filtrad los datos, para extraer solo aquellos productos que tengan uno o más dígitos en su titulo. ¿Cuántos productos hay?
df_unido_digitos = df_unido[df_unido["title_orig"].str.contains("\d", regex = True, na = False)]
df_unido_digitos.shape
# Hay 581 productos.

#¡HEMOS TERMINADO!
```
